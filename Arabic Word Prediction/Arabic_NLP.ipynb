```python
import os
import re
THRESHOLD = 0.0
from collections import defaultdict
```

```python
inputPath = r'articlesCulture'
nonForamtedSentences = []
for filename in os.listdir(inputPath):
    if (filename == 'culturecjan306.txt') :
        break
    f = open(inputPath + '\\' + filename, 'r',encoding = 'utf-8')
    content = f.read()
    list_ar = content.split(sep ='\n')
    nonForamtedSentences.append(list_ar)
```

```python
training_corpus = []
for sentence in nonForamtedSentences:
    text = ' '.join(sentence)
    text = re.sub(r'[^\u0627-\u064a\s]','', text)
    sentence = text.strip().split()
    training_corpus.append(sentence)
```

```python
numberOfWords = 0
for sentence in training_corpus:
    numberOfWords += len(sentence)
numberOfWords
```

```python
unigramModel = {}

for sentence in training_corpus:
    for word in sentence :
        if word not in unigramModel :
            unigramModel[word]=1
        else:
            unigramModel[word]+=1

for word in unigramModel:
    unigramModel[word] /= numberOfWords
```

```python
from nltk import  bigrams

bigramCounter = defaultdict(lambda: defaultdict(lambda: 0))
bigramProbability = defaultdict(lambda: defaultdict(lambda: 0))
 

for sentence in training_corpus:
    for w1, w2 in bigrams(sentence, pad_right=True, pad_left=True):
        bigramCounter[w1][w2] += 1
        
for w1 in bigramCounter:
        totalUnigramCount = float(sum(bigramCounter[w1].values()))
        for w2 in bigramCounter[w1]:
            bigramProbability[w1][w2] = bigramCounter[w1][w2] / totalUnigramCount
```

```python
from nltk import  trigrams
trigramCounter = defaultdict(lambda: defaultdict(lambda: 0))
trigramProbability = defaultdict(lambda: defaultdict(lambda: 0))
 

for sentence in training_corpus:
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        trigramCounter[(w1, w2)][w3] += 1
```

```python
with open('outpute.txt', 'w+',encoding = 'utf-8') as f:
    for w1_w2 in trigramCounter:
        totalBigramCount = float(sum(trigramCounter[w1_w2].values()))
        for w3 in trigramCounter[w1_w2]:
            trigramProbability[w1_w2][w3] = trigramCounter[w1_w2][w3] / totalBigramCount
            if (trigramProbability[w1_w2][w3] > THRESHOLD) :
            f.write(str(w3) +" | " + str(w1_w2) +  " = " + str( trigramProbability[w1_w2][w3])  + '\n')
        
```

```python
def handle(x):
    lst=[]
    lst=re.split('\s+', x)
    s2 = ""
    s3 = ""
    
    if (len(lst) >= 2):
        s2=lst[-2]
        s3=lst[-1]
        
    elif ((len(lst)  == 1) and (lst[0] != '')):
          s3=lst[-1]
    else:
          s3 = "خطأ"
               
    results =[]
    finalResult=[]
    
    if (len(trigramCounter[(s2,s3)]) > 0):
        for W3 in trigramCounter[(s2,s3)]:
            temp=[W3,trigramCounter[(s2,s3)][W3]]
            results.append(temp)
        results.sort(key=lambda k: k[1] , reverse=True)
        for i in results:
            finalResult.append(i[0])
    
    elif (len(bigramCounter[s3]) > 0):
        results =[]
        finalResult=[]
        for W3 in bigramCounter[s3]:
            temp=[W3,bigramCounter[s3][W3]]
            results.append(temp)
        results.sort(key=lambda k: k[1] , reverse=True)
        for i in results:
            finalResult.append(i[0])
    
    elif (s3 in unigramModel) :
        finalResult = unigramModel.sort(key=lambda k: k[1] , reverse=True)[10]
    else:
        finalResult = s3
        
          
    if(len(finalResult)>10):
        finalResult = finalResult[:10]
            
    return  finalResult
```

```python

from tkinter import  ttk
from tkinter import  *


def func(*args):
    x=combobox.get()
    combobox['values'] = handle(x)

    

root =Tk()

combobox = ttk.Combobox(root,width=20)
combobox.focus()
combobox.pack()
root.bind("<Return>",func)
root.geometry("450x450")
root.mainloop()
```
