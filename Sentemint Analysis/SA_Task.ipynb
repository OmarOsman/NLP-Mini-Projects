```python
import pandas as pd
import random as r
import nltk
import re
from nltk.corpus import stopwords
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.linear_model import LogisticRegression

```

```python
f = open('TrainData.txt', 'r', encoding='utf-8')
content = f.read()
train = content.split(sep='\n')
```

```python
f = open('TrainDataLabel.txt', 'r', encoding='utf-8')
content = f.read()
trainLabel = content.split(sep='\n')
```

```python
f = open('TestData.txt', 'r', encoding='utf-8')
content = f.read()
test = content.split(sep='\n')
```

```python
f = open('TestDataLabel.txt', 'r', encoding='utf-8')
content = f.read()
testLabel = content.split(sep='\n')
```

```python
from nltk.stem.porter import *

def preprocess (sentence):  
    letters_only = re.sub("[^a-zA-Z]", " ", sentence) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops]
    meaningful_words = [w for w in meaningful_words if len(w)>2]
    stemmer = PorterStemmer()
    meaningful_words = [stemmer.stem(meaningful) for meaningful in meaningful_words]
    return( " ".join( meaningful_words ))   
```

```python
preprocessedTrain = []
for s in train :
    preprocessedTrain.append(preprocess(s))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = "word",   
                             tokenizer = None,    
                             preprocessor = None, 
                             stop_words = None,   
                             max_features = len(set(preprocessedTrain))) 

train_data_features = vectorizer.fit_transform(preprocessedTrain)
train_data_features = train_data_features.toarray()
```

```python
vocab = vectorizer.get_feature_names()
#print (vocab)
```

```python
print (train_data_features[50])
```

```python
import numpy as np
dist = np.sum(train_data_features, axis=0)
for tag, count in zip(vocab, dist):
    print (count, tag)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 205) 
forest = forest.fit(train_data_features, trainLabel)
```

```python
preprocessedTest = []
for s in test :
    preprocessedTest.append(preprocess(s))
```

```python
test_data_features = vectorizer.transform(preprocessedTest)
test_data_features = test_data_features.toarray()
result = forest.predict(test_data_features)
```

```python
test_data_features.shape
```

```python
def accuracy(resultList,testLabel) :
    tr = 0
    for i in range (len(testLabel)):
        if (resultList[i] == testLabel[i]):
            tr+=1       
    score = tr / len(testLabel) * 100
    return score
```

```python
print ("Random Forest Accuracy : " + str(accuracy(list(result),testLabel)))
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(train_data_features, trainLabel)
predict = clf.predict(test_data_features)
```

```python
print ("Gaussian Naive Bayes : " + str(accuracy(testLabel, predict)))
```

```python
from sklearn import svm
svmClf = svm.SVC(decision_function_shape='ovo',kernel='linear') 
svmClf = svmClf.fit(train_data_features, trainLabel)
```

```python
test_data_features = vectorizer.transform(preprocessedTest)
test_data_features = test_data_features.toarray()
result = svmClf.predict(test_data_features)
```

```python
print ("SVM : " + str(accuracy(testLabel, result)))
```
